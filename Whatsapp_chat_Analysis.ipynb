```python
import re
import pandas as pd
```

```python
f = open('WhatsApp Chat with Priyanshu Gupta.txt', 'r', encoding='utf-8')
```

```python
data = f.read()
```

```python
print(data)
```

```python
print(type(data))
```

```python
pattern = '\d{1,2}/\d{1,2}/\d{2,4},\s\d{1,2}:\d{2}\s(?:am|pm|AM|PM)\s-\s'
```

```python
messages = re.split(pattern, data)[1:]
len(messages)
print(messages)
```

```python
dates = re.findall(pattern, data)
dates
```

```python
df = pd.DataFrame({'user_message':messages, 'message_date': dates})
df['message_date'] = pd.to_datetime(df['message_date'], format='%d/%m/%y, %I:%M %p - ')
df.rename(columns={'message_date': 'date'}, inplace=True)
df.head()
```

```python
df.shape
```

```python
df.tail()
```

```python
# Separating users and messages
# Full Expression Explanation:
#([\w\W]+?) :
#Captures any sequence of characters, stopping at the shortest match due to the lazy ?.
#: :
#Ensures the match stops just before a colon.
#\s :
#Ensures thereâ€™s a space after the colon.
users = []
messages = []
for message in df['user_message']:
    entry = re.split('([\w\W]+?):\s', message)
    if entry[1:]:
        users.append(entry[1])
        messages.append(entry[2])
    else:
        users.append('group_notification')
        messages.append(entry[0])
        
df['user'] = users
df['message'] = messages
df.drop(columns=['user_message'], inplace=True)

df.head()
```

```python
df['year'] = df['date'].dt.year
```

```python
df['day'] = df['date'].dt.day
df['hour'] = df['date'].dt.hour
df['minute'] = df['date'].dt.minute
```

```python
temp = df[df['user'] != 'group_notification']
temp = temp[temp['message'] != '<Media omitted>\n']
```

```python
f = open('stop_hinglish.txt', 'r')
stop_words = f.read()
print(stop_words)
```

```python
words = []
for message in temp['message']:
    for word in message.lower().split():
        if word not in stop_words:
            words.append(word)
```

```python
from collections import Counter
pd.DataFrame(Counter(words).most_common(20))
```
